```python
# pd.read_csv('Titanic Data')
train_d = pd.read_csv('/kaggle/input/titanic/train.csv')
#train_d = train_d.dropna(axis=0)
train_d.describe()


# Survival Statistics of Training Data:
# Dead: 61.6 %
# Alive: 38.4%
```

```python
test_d = pd.read_csv("/kaggle/input/titanic/test.csv")
# More Descriptive Analysis of Titanic Testing data using Pandas Library
# Drops Missing Values if necessary : test_d.dropna(axis = 0)
# test_d = test_d.dropna(axis = 0)
#test_d.head()
print(type(test_d))
test_d = test_d.fillna(test_d["Fare"].mean())
test_d.describe()
```

```python
women = train_d.loc[train_d.Sex == 'female']["Survived"]
#Makes list of females and their respective survival values
print(women)
rate_women = sum (women) / len (women) # sum (women) = # Women who survived as 1 is assigned to women who survive
print("% of women who survived:", rate_women)
```

```python
men = train_d.loc[train_d.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
# Ticket Classes: Upper, Second, Third
first_class = train_d.loc[train_d.Pclass == 1]["Survived"]
rate_first_c = sum (first_class) / len (first_class)
print("% of first class passengers who survived:", rate_first_c)

# Major Correlation between being a member of first class and survival rate
```

```python
sec_class = train_d.loc[train_d.Pclass == 2]["Survived"]
rate_sec_c = sum (sec_class) / len (sec_class)
print("% of second class passengers who survived:", rate_sec_c)
```

```python
third_class = train_d.loc[train_d.Pclass == 3]["Survived"]
rate_third_c = sum (third_class) / len (third_class)
print("% of third class passengers who survived:", rate_third_c)
```

```python
# Hence as class decreases, survival rate decreases
```

```python
# Age Mathematical Analysis
# Mean age = 29.7 years
max_age = 0
rng_ages = [10,20,30,40,50,60,70,80]
for i in rng_ages:
    age_t1 = train_d.loc[train_d.Age < i]["Survived"]
    rate_age = sum(age_t1)/len(age_t1)
    if(rate_age > max_age):
        max_age = rate_age
        z = i
print(type(age_t1))
print(z)
# 61.3 % of passengers less than 10 survived 
# 66.7 % of passengers less than 9 sruvived

print("% of passengers below the age of who survived:", max_age)
```

```python
# Fare Mathematical Analysis
# Note : 32.2 used as it is the mean fare
fare = train_d.loc[train_d.Fare < 51.23]["Survived"]
fare_t = train_d.loc[train_d.Fare > 51.23]["Survived"]
rate_fare = sum (fare) / len (fare)
rate_fare_t = sum (fare_t) / len (fare_t)
print("% of passengers who survived with fare below 51.23: ", rate_fare)
print("% of passengers who survived with fare above 51.23 : ", rate_fare_t)
# Overall Survival Rate: 38.4% 
# Not enough agreement to establish strong correlation
```

```python
# SibSp Mathematical Analysis
for i in range(0,9):
        if i not in [6,7]: 
            sibsp = train_d.loc[train_d.SibSp == i]["Survived"]
            rate_sibsp = sum (sibsp) / len (sibsp)
            print("% of passengers who survived when sibsp equal to", i, " : ", rate_sibsp)
# No passengers in training data with 6 or 7 Siblings
# No relation found between sibsp and survival rate
```

```python
# Parch Mathematical Analysis
for i in range(0,7):
        parch = train_d.loc[train_d.Parch == i]["Survived"]
        rate_parch = sum (parch) / len (parch)
        print("% of passengers who survived when parch equal to", i, " : ", rate_parch)
# No relation found between parch and survival rate
# However, when parch = 2, survival rate considerably higher and many values to support this
```

```python
from sklearn.ensemble import RandomForestClassifier
# Imports Random Forest Model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import itertools

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.

y = train_d["Survived"] # Denotes array of survivial values (0 or 1) matched to each individual

#Test embarked, fare, sibSp, parch
features_list = ["Embarked", "Fare", "SibSp", "Parch"]
features = ["Pclass", "Sex"]


#Features Size 3:
for x in features_list:
    fx = ["Pclass", "Sex"]
    fx.append(x)
    X = pd.get_dummies(train_d[fx])
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
    model = RandomForestClassifier(n_estimators = 100, max_depth = 5 , random_state =1 )
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    print(x, "_MAE: ", mean_absolute_error(val_y, predictions))

#Features Size 4:
        
ft_c = itertools.combinations(features_list, 2)
ft_c = list(ft_c)

for z in ft_c:
    z = list(z)
    fz = ["Pclass", "Sex"]
    fz.extend(z)
    X = pd.get_dummies(train_d[fz])
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
    model = RandomForestClassifier(n_estimators = 100, max_depth = 5 , random_state = 1 )
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    print(z, "_MAE: ", mean_absolute_error(val_y, predictions))
    
#Features Size 5:
        
ft_c = itertools.combinations(features_list, 3)
ft_c = list(ft_c)

for z in ft_c:
    z = list(z)
    fz = ["Pclass", "Sex"]
    fz.extend(z)
    X = pd.get_dummies(train_d[fz])
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
    model = RandomForestClassifier(n_estimators = 100, max_depth = 5 , random_state = 1 )
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    print(z, "_MAE: ", mean_absolute_error(val_y, predictions))


# Feature Size 6

features_list.extend(features)
X = pd.get_dummies(train_d[features_list])
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
model = RandomForestClassifier(n_estimators = 100, max_depth = 5 , random_state = 1 )
model.fit(train_X, train_y)
predictions = model.predict(val_X)

print("Size 6: ", mean_absolute_error(val_y, predictions))

# Features Size 2: 
X = pd.get_dummies(train_d[features])
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
# Fit on testing Data variables X and y
model.fit(train_X, train_y)
predictions = model.predict(val_X)
print(mean_absolute_error(val_y, predictions))

#consider adding cabin to features

# Note: Assumed that there is direct correlation between mean absolute error of predictions on validation data and val_y and 

features_final = ["Pclass", "Sex", "Embarked", "Fare"]
X = pd.get_dummies(train_d[features_final])
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# FINAL MODEL DESIGN 
#.get_dummies 
# n_estimators = number of trees in the foreset

# Estimator Tuning : 
for i in [5, 10, 15, 16, 17, 18, 19, 20, 25, 30]:
    model_lf = RandomForestClassifier(n_estimators = i, max_depth = 5, random_state = 1)
    model_lf.fit(train_X, train_y)
    prdc = model_lf.predict(val_X)
    print("N Estimators Size : ", i, " : ", mean_absolute_error(val_y, prdc))

# Baseline MAE = 0.19730941704035873
# MAE_Size : 2 = 0.21973094170403587 Features: ["Pclass", "Sex"]
# Best MAE Found with n_estimators = 100, max_depth = 5 : 
# ['Embarked', 'Fare'] _MAE:  0.17937219730941703
# Leaf Tuned MAE : 0.17488789237668162


# Optimal Leaf Size found to be from 0 to 17; using validation data and MAE as success metric
final_model = RandomForestClassifier(n_estimators = 17, max_depth = 5, random_state = 1)
final_model.fit(X, y)
test_X = pd.get_dummies(test_d[features_final])
final_predictions = final_model.predict(test_X)

output = pd.DataFrame({'PassengerId': test_d.PassengerId, 'Survived': final_predictions})
output.to_csv('submission_titanic_leaf_tuned.csv', index = False)

print("Your submission was successfully saved!")
```
